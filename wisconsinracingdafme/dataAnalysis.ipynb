```python
import pandas as pd
import matplotlib.pyplot as plt
import time
globaldata = pd.read_csv("parsedDF.csv", low_memory=False)
```

```python
def basicanalysis(data):
    # print(data.head())
    """
    Average Speed
    Max AY (Acceleration Y)
    Max AX(Acceleration X)
    Name and Date (I think you can pull this from mdf metadata, unsure if its available)
    For E Car -> Duration of time in HV, C Car -> Duration of time above 1000 RPM
    Num Motors Active (E Car)

    """
    t0 = time.time()
    basicoutput = {}
    data = data[1:]
    # print(data.dtypes)
    # data = data[pd.to_numeric(data['VehicleSpeed'], errors='coerce').notnull()]
    # data["VehicleSpeed"] = pd.to_numeric(data.VehicleSpeed, errors='coerce')
    data = data.apply(pd.to_numeric, errors='ignore')
    # print(data.dtypes)
    avgspeed = pd.Series(data["VehicleSpeed"]).mean()
    avgTrueSpeed = pd.Series(data[data["VehicleSpeed"] > 1]["VehicleSpeed"]).mean()
    max_accel_Y = data["AccelY"].max()
    max_accel_X = data["AccelX"].max()
    fig, ax = plt.subplots()
    fig2, ax2 = plt.subplots()

    basicoutput = {
        "Average Speed": avgspeed, 
        "Average True Speed": avgTrueSpeed, 
        "Max AY (Acceleration Y)": max_accel_Y, 
        "Max AX (Acceleration X)": max_accel_X
    }

    # data["VehicleSpeed"].plot()
    # ax2 = pd.Series(data["RRMotorSpeed"]).plot.line()
    # data["RRMotorSpeed"].plot()
    # data["RLMotorSpeed"].plot()
    ax.plot(data["VehicleSpeed"])
    # plt.show()
    ax2.plot(data["RRMotorSpeed"])
    ax2.plot(data["RLMotorSpeed"])

    t1 = time.time()
    print(f"Time to process: {(t1-t0)}" )
    return basicoutput

basicanalysis(globaldata)
```
